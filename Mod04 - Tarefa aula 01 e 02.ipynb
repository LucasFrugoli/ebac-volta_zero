```python
# seu código aqui
df['media'] = (df['x1'] + df['x2'] + df['x3'])/3
df
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# seu código aqui
df['log_med'] = None
x1 = []

for i in df['media']:
    x = math.log1p(i)
    x1.append(x)
df['log_med'] = x1
df
```
